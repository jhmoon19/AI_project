## css_selector

- bs4의 select()와 동일

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome("chromedriver")

driver.get(url)
print(driver, '\n')

selected = driver.find_element(by = By.CSS_SELECTOR, value = "div.home div.p")
# 클래스가 home인 div클래스의 자손 중 클래스가 p인 div 클래스 
print(selected, '\n')
print(selected.tag_name, '\n') # == [soup].name 
print(selected.text)
```

- 메인 페이지에서 돔트리를 불러오고 다른 페이지로 넘어가면 그 전에 가져온 페이지는 사용할 수 없게 됨

- 따라서 click을 페이지 이동 용도로 사용하는 것은 가급적 피하는 것이 좋음

- 페이지 변화 없이 페이지 내에서 데이터가 변화되는 경우에 사용하는 것을 권장

```python
url = "https://pjt3591oo.github.io"

driver = webdriver.Chrome("chromedriver")

driver.get(url)

selected_tags_a = driver.find_elements(By.CSS_SELECTOR, "a")

for i in selected_tags_a:
    time.sleep(2)
    try:
        print(i.text, i.tag_name)
        i.click()
    except:
        print("페이지가 이동되어 오류가 났습니다.")
        break
```
