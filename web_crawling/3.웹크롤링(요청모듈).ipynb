- 없는 페이지를 요청해도 에러를 띄운다

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://davelee-fun.github.io/")
print(res, '\n')
soup = BeautifulSoup(res.content, 'html.parser')
print(res.content, '\n')
print(soup, '\n')
data = soup.select('h4.card-text')
print(data, '\n')

for item in data:
    print(item.get_text().strip())
```
