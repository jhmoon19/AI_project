# 과제 1-2. 티스토리 크롤링

- url : https://ai-dev.tistory.com/2


- 테이블 내용 수집
    - 상품, 색상, 가격, 셔츠1, 빨강, 20000, ......

```python
url = "https://ai-dev.tistory.com/2"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')
print(soup)
```

```python
a = soup.find('meta', property='og:description')['content']

a[a.index('상품'):a.index('이번에는')-1]
```

```python
# 다른 방법
table_text = soup.select("div.tt_article_useless_p_margin table tbody tr td")

for i in table_text:
    print(i.string)
```

# 과제 2. 네이버 분야별 주요뉴스 수집

- url : https://news.naver.com/main/officeList.naver
- 화면 오른쪽 아래 분야별 주요뉴스의 기사 제목 수집

```python
url = "https://news.naver.com/main/officeList.naver"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')

print(soup)
```

```python
soup.find_all('ul', class_='list_txt')
```

```python
for i in soup.find_all('ul', class_='list_txt'):
    for j in i.find_all('a', class_="nclicks(rig.secteco)"):
        print(j.text)
```

```python
# 다른 방법

for i in soup.select("div.section div.classfy ul li"):
    print(i.text)
```

# 과제 3. 다음 영화 랭킹 수집

- url : https://movie.daum.net/ranking/reservation


- 예매순위의 영화제목 수집

```python
url = "https://movie.daum.net/ranking/reservation"
html = urlopen(url)
soup = BeautifulSoup(html, 'lxml')

print(soup)
```

```python
for i in soup.find_all('div', class_="poster_movie"):
    print(i.img['alt'])
```

```python
for i in soup.find_all('div', class_="poster_movie"):
    print(i.img)
```

```python
for i in soup.find_all('div', class_="poster_movie"):
    print(i.find('img')['alt'])
```

```python
for i in soup.find_all('div', class_="poster_movie"):
    print(i.img.attrs['alt'])
```

```python
for i in soup.find_all('img', class_="img_thumb"):
    print(i['alt'])
```

```python
driver = webdriver.Chrome("chromedriver")
driver.get(url)

selected_tags = driver.find_elements(By.CSS_SELECTOR, "img.img_thumb")

for i in selected_tags:
    print(i.contents)
```
