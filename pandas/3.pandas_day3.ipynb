## 결측치 대체

```python
ebola.iloc[:10, :5]
```

```python
ebola.fillna(0).iloc[:10, :5]
```

```python
ebola.fillna("미친놈").iloc[:10, :5]
```

```python
# ffill : 누락값이 나타나기 전의 값으로 채워넣기
ebola.fillna(method = "ffill").iloc[:10, :5]
```

```python
# bfill : 누락값 이후 첫 번째 값으로 앞쪽의 누락값이 변경
ebola.fillna(method = "bfill").iloc[:10, :5]
```

```python
# interpolate() : 누락값 양쪽값의 중간값으로 처리
# 데이터가 일정한 간격을 유지하고 있을 때 유용함
ebola.interpolate().iloc[:10, :5]
```

1) ebola.fillna(0).iloc[:10,:5] - 0.0으로

2) ebola.fillna(method=ffill).iloc[:10, :5] - 앞의 값 

3) ebola.fillna(method=bfill).iloc[:10, :5] - 뒤의 값

4) ebola.interpolate().iloc[:10, :5] - 앞뒤 중간값
