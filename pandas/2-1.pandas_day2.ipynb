```python
data_dict = {"A" : "n1", "B" : "n2", "C" : "n3", "D" : "n4"}
```

```python
df1.append(data_dict, ignore_index = True)
# ignore_index =True: 원래의 인덱스를 무시하고, 자연스레 이어붙임 
```

```python
pd.DataFrame(data_dict, index = [0]) # 모든 value가 스칼라여서, index=[0] 지정 필요 --> 아니면 오류 
```

```python
pd.DataFrame({'A':['n1'], 'B':['n2']}) # 아니면 이런식으로 
```

```python
df1.append(pd.DataFrame(data_dict, index = [0]))
# ignore_index=True 안할 시, index를 그대로 0으로 이어붙임 
```

- 딕셔너리는 인덱스가 없기 때문에 ignore_index가 없으면 에러
- ignore_index = True로 하면 데이터를 연결한 다음 인덱스를 다시 지정

```python
concat_df1 = pd.concat([df1, df2, df3], ignore_index = True) 
# 디폴트: axis=0 (행방향-위아래방향 연결)
concat_df1
```

## 열 방향으로 연결

```python
col_concat = pd.concat([df1, df2, df3], axis = 1)
# ignore_index 안해줬기 때문에 열명 그대로 ABCABC..
col_concat
```

```python
col_concat["A"] # 열명이 같아서 df로 모두 
```

```python
col_concat["new_col_list"] = ["n1", "n2", "n3", "n4"]
col_concat
```

```python
# 열이름을 다시 지정
pd.concat([df1, df2, df3], axis = 1, ignore_index = True)
# 좌우로 이어붙이는 경우엔, ignore_index 하면, 
# 원래의 열명을 무시하고, 숫자 0~ 로 열명이 설정됨 
```

## 공통 열과 공통 인덱스만 연결하기

```python
df1.columns = ["A", "B", "C", "D"]
df2.columns = ["E", "F", "G", "H"]
df3.columns = ["A", "C", "F", "H"]
```

```python
df3
```

```python
row_concat = pd.concat([df1, df2, df3])
row_concat
```

```python
# 공통 열만 연결 (join='inner')
pd.concat([df1, df3], join = "inner")
```

```python
pd.concat([df1, df3], join = "inner", ignore_index=True)
```

```python
pd.concat([df1, df2, df3], join = "inner")
# 세 df 모두 겹치는 열이 없기 때문에 행번호만..
```

```python
# join='outer': 겹치지 않은 열 데이터는 NaN으로 표시
pd.concat([df1, df3], join='outer')
```

# merge

```python
df1 = pd.DataFrame({"key" : list("bbacaab"), "data1" : range(7)})
df2 = pd.DataFrame({"key" : list("abd"), "data2" : range(3)})
```

```python
df1
```

```python
df2
```

```python
# key 열 기준으로 데이터 병합 (merge)
pd.merge(df1, df2, on = "key")
# 겹치지 않은 key는 버려짐 
```

```python
# merge 예제2
exam = pd.read_csv("./data/csv_exam.csv")
```

```python
exam.head()
```

```python
df_name = pd.DataFrame({"class" : [1, 2, 3, 4, 5],
                        "teacher" : ["kim", "lee", "park", "choi", "jung"]})
```

```python
df_name.head()
```

```python
# 해당 class에 선생님 이름 정보를 추가하고 싶다 --> concat X, merge O
pd.merge(exam, df_name, on = "class")
```

# 누락값 처리

```python
from numpy import NaN
```

```python
print(pd.isnull(NaN))
```

```python
print(NaN == True)
print(NaN == False)
print(NaN == 0)
print(NaN == '')
print(NaN == NaN)
# null값 NaN은 False, 0, '' 중 그무엇도 아니고
# 심지어 NaN 과 NaN이 같다고 할 수도 없음
```
